Menerapkan agregasi pada sejumlah kolom dengan agg()

```python
import pandas as pd
print(pd.__version__)
```

```python
df = pd.read_csv('C:/Users/Rosi Rachel/Desktop/Data Frame/titanicfull.csv')
df.head()
```

```python
df.groupby('pclass').agg({'pclass':'count', 'age':['mean', 'max'], 'survived':'mean'})
```

```python
df.groupby('pclass').agg(n_pass=('pclass', 'count'), avg_age=('age', 'mean'), 
                        max_age=('age', 'max'), survival_rate=('survived', 'mean'))
```
